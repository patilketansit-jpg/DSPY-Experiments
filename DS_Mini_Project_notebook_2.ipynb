# Titanic Dataset – Exploratory Data Analysis (EDA)

In this notebook, we will use the **filtered Titanic dataset** from Notebook 1, which has already undergone necessary data cleaning and preprocessing. This notebook will perform **Exploratory Data Analysis (EDA)** on the cleaned data to extract meaningful patterns, visualize relationships, and prepare for the machine learning model.

We will perform:

- **2 Non-Graphical Analyses** (Univariate and Multivariate)
- **2 Graphical Analyses** (Univariate and Multivariate)
- **A mix of univariate and multivariate analyses** with clear observations and conclusions


Non-Graphical EDA (Text/Stats)

Before starting the analysis, let's load the cleaned data from Notebook 1.

```python
import pandas as pd

# Load the filtered dataset from Notebook 1
df = pd.read_csv('titanic_cleaned.csv')

# Check the first few rows to ensure the data has been loaded correctly
df.head()
```

1. Univariate Analysis – Survival Rate

Information Sought:
We are interested in examining the proportion of passengers who survived vs. those who did not. This will help us understand the overall survival rate of the passengers.

Type of EDA:
This is a univariate analysis focusing on a single variable (Survived).

```python
# Percentage of passengers who survived vs. who didn't
df['Survived'].value_counts(normalize=True)
```

Observation: Around 38% of passengers survived, while 62% did not.

Concluding Remarks: This result shows a relatively low survival rate, with many passengers perishing in the disaster. This baseline statistic is helpful for future comparisons.



2. Multivariate Analysis – Survival by Sex

Information Sought:
We want to explore how the survival rate varies based on gender. This analysis helps us understand the relationship between sex and survival.

Type of EDA:
This is a multivariate analysis examining the relationship between two variables (Sex and Survived).

```python
# Average survival rate by gender
df.groupby('Sex')['Survived'].mean()
```

```python
# Load original dataset
original_df = pd.read_csv('titanic.csv')

# Add back the Sex column if needed
df['Sex'] = original_df['Sex']
```

```python
print(df.groupby('Sex')['Survived'].mean())
```

Observation:Females had a much higher survival rate (74%) compared to males (~19%).

Concluding Remarks:
Gender was a significant factor in survival, with females having a higher chance of survival compared to males.

Graphical EDA

 3. Univariate Analysis – Age Distribution

Information Sought:
We aim to examine the distribution of passenger ages. This will help us determine whether there are any noticeable patterns or age-related trends in the data.

Type of EDA:
This is a univariate graphical analysis, where we plot the distribution of Age.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting Age distribution
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

Observation:

Most passengers were aged between 20 and 40 years, with a peak around the mid-30s.

Concluding Remarks:
Age does not show a strong skew, with a relatively normal distribution. While age itself doesn’t seem to be a decisive factor for survival, younger passengers were more frequent on the Titanic.



4. Multivariate Analysis – Survival by Pclass and Sex

Information Sought:
We want to understand how passenger class (Pclass) and sex affect the survival rate.

Type of EDA:
This is a multivariate graphical analysis, where we explore the relationship between three variables: Pclass, Sex, and Survived

```python
# Survival rate by class and gender
plt.figure(figsize=(8, 6))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df)
plt.title('Survival Rate by Pclass and Sex')
plt.ylabel('Survival Rate')
plt.xlabel('Passenger Class')
plt.show()
```

Observation:1st class females had the highest survival rate, while 3rd class males had the lowest survival rate.There is a strong relationship between class and survival, with higher classes generally having higher survival rates, particularly for females.

Concluding Remarks:
Passenger class and gender were significant factors in survival. Passengers in 1st class, especially females, had the highest chance of survival.

Summary:

Women and passengers in higher classes (especially 1st class) had a significantly higher chance of survival.

The majority of passengers were between 20 and 40 years old, but age alone didn’t strongly predict survival.


# 🧾 Conclusion

We observed:
- Females and 1st class passengers had higher chances of survival.
- Most passengers were in 3rd class and paid lower fares.
- Age distribution was fairly centered around 29 years.
- Strong influence of gender and class on survival.

This EDA provides strong insight for building an ML model in the next notebook.
